```python
import pandas as pd
import numpy as np

'''
This is a short demo on looking at the columns of U from the SVD of a matrix, A = US(V.T) 

A is an n x m matrix where n = number of data points and m = number of features
'''
```

```python
#data description
df = pd.read_csv("food-consumption.csv")
#change int64 columns to float64 columns
for (col, dat) in df.iteritems():
    if df[col].dtype=='int64':
        df[col] = df[col].astype('float64')
        
#replace missing data with 0. not always the best solution
df = df.replace(np.nan,0.0)

df
```

```python
#remove string headers for country names
df = df.iloc[:,1:]
```

# Null Space, Lossy Compression, and U(U.T) for different ranks

```python
food_mat = df.to_numpy(dtype=np.float64)# food_mat = A
print("shape of data: ", food_mat.shape)
col_avg = np.reshape(food_mat.mean(1), (food_mat.shape[0],1))
col_avg = np.tile(col_avg, (1, food_mat.shape[1]))

food_mat = food_mat - col_avg
```

```python
print("Shapes with \' full matrices\' off: ", )
U, S, Vt = np.linalg.svd(food_mat, full_matrices=False)
rank = S.shape[0]
print("U shape: ", U.shape)
print("S shape: ", S.shape)
print("Vt shape: ", Vt.shape)
print("")
print("Shapes with \' full matrices\' on: ", )
U, S, Vt = np.linalg.svd(food_mat, full_matrices=True)
rank = S.shape[0]
print("U shape: ", U.shape)
print("S shape: ", S.shape)
print("Vt shape: ", Vt.shape)
```

```python
#what are the extra rows in Vt?
#they are the basis for the null space of A
```

```python
#exploration of null vec of data is N x M
print("rank: ", rank)
null_vecs = Vt[rank:].T
null_vec_coeffs = np.random.random((null_vecs.shape[1],))

#create a random null vector by random linear combination of null basis
random_null_vec = np.matmul(null_vecs, null_vec_coeffs)


print("result of multiplying A by a random null vector: ")
print(np.matmul(food_mat, random_null_vec))
```

```python
'''
Recall that y = U(U.T)x is the projection of x onto the span of orthonormal basis, U

Projection_coefficients = (U.T)x

Projection = U(Projection_coefficients) = U(U.T)x

Data compression
'''

#num cols of U = 16
k = 1
smallU = U[:,:k]
print("matrix U shape: ", U.shape)
print("matrix smallU shape: ", smallU.shape)
x = np.expand_dims(food_mat[:,0],axis=1)
#x = np.random.random((food_mat[:,0].shape))
print("x shape: ", x.shape)
print("")
print("first column of centered original data: ", x.T)
mat_projection = np.matmul(smallU.T, x)
mat_projection = np.matmul(smallU, mat_projection)

print("")
print("coefs: ", np.matmul(smallU.T,x).T)

print("")
print("centered Projection (U x coefs): ", mat_projection.T)

print("")
print("distance between original and projection: ", np.linalg.norm(mat_projection - x))
```

```python
'''
How many columns of U to use? i.e. how many column vectors of U to use for a reasonable reconstruction?

Take a look at S from SVD
'''
k = 1
print(S)
variance_accounted_for = 0
total_variance = 0
for i in range(len(S)):
    if i < k:
        variance_accounted_for += S[i]**2
    total_variance += S[i]**2
print("total variance accounted for using k = ", k, " columns of U: ", variance_accounted_for/total_variance)
```

```python
'''
Interesting property: if I look at U (or V) from SVD, and consider U(U.T), we notice the following

U(U.T) = identity
'''

#look at submatrices of U which take the first k columns and compute (U.T)U
k = 16
smallU =U[:,:k]
print("rank: ", np.linalg.matrix_rank(smallU))
print("")
#print(np.matmul(smallU,smallU.T))

print("difference between smallU(smallU.T) and identity: ", np.linalg.norm(np.matmul(smallU,smallU.T) - np.eye(16)))
```

# Matrix Completion

Consider the following matrix that is missing entries: how can we fill in the missing entries in a meaningful way?

```python
df = pd.read_csv("food-consumption.csv")
print("number of NaNs: ", df.isna().sum().sum())
df
```

## Very basic method: using SVD and low rank approximation
![basic algorithm](4lowrankapproximation.PNG)

```python
#replace NaN with something to so computation can be done, here I replaced with 0's

for (col, dat) in df.iteritems():
    if df[col].dtype=='int64':
        df[col] = df[col].astype('float64')
        
#replace missing data with 0. not always the best solution
df = df.replace(np.nan,0.0)
df = df.iloc[:,1:]
```

```python
food_mat = df.to_numpy(dtype=np.float64)# food_mat = A
print("shape of data: ", food_mat.shape)
col_avg = np.reshape(food_mat.mean(1), (food_mat.shape[0],1))
col_avg_mat = np.tile(col_avg, (1, food_mat.shape[1]))

food_mat = food_mat - col_avg_mat
```

```python
print("column with missing data: ", food_mat[:,3]+np.squeeze(col_avg))
```

```python
k = 10
U, S, Vt = np.linalg.svd(food_mat, full_matrices = False)
print("U shape:", U[:,:k].shape)
print("S shape: ", S[:k].shape)
print("Vt shape: ", Vt[:k,:].shape)
```

```python
A_k = np.matmul(np.diag(S[:k]), Vt[:k,:])
A_k = np.matmul(U[:,:k], A_k)

print("column with missing data: ", A_k[:,3]+np.squeeze(col_avg))
```
